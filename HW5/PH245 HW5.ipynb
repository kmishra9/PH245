## Problem 1

```r
library(MASS)
library(mda)
library(class)
library(ggplot2)
library(tree)

# Loading Data 
breastCancer = read.table(file="Data-HW5-breastcancer.dat", 
                   header=TRUE, 
                   quote="", 
                   sep=""
                  )

# Displaying Data
print(paste("Number of Patients:", nrow(breastCancer)))
head(breastCancer)
```

#### Problem 1A

```r
# Partitioning the data (trainingSet=400 patients, testSet=169 patients)

set.seed(1000)

trainSize = 400
testSize  = 169
sampleSize = trainSize+testSize

trainIndices = sample(seq(1:sampleSize), size=trainSize)

trainData = breastCancer[trainIndices,]
testData = breastCancer[-trainIndices,]

# Ensuring our results match our expectations
print(paste("Training data, number of rows:", nrow(trainData)))
head(trainData)
print(paste("Test data, number of rows:", nrow(testData)))
head(testData)
```

#### Problem 1B

```r
DATrainer = function(train, test, DAType) {
    DAFit = DAType(formula=y~., data=train)
    predictions = predict(DAFit, test)$class
    misclassificationTestErrorRate = sum(predictions != test[,1])/length(predictions)
    return(misclassificationTestErrorRate)
}


# LDA
print(paste("LDA Misclassification Rate: ", DATrainer(train=trainData, test=testData, DAType=lda)))

# QDA
print(paste("QDA Misclassification Rate: ", DATrainer(train=trainData, test=testData, DAType=qda)))
```

```r
# MDA

MDATrainer = function(train, test, subclasses) {
    MDAFit = mda(formula=y~., data=train, subclasses=subclasses)
    predictions = predict(MDAFit, test[,-1])
    return( sum(predictions!=test[,1])/length(predictions) )
}

print(paste("MDA Misclassification Rate: ", MDATrainer(train=trainData, test=testData, subclasses=c(5,5))))
```

```r
# CART

treeFit = tree(formula=as.factor(y)~., data=trainData)
predictions = predict(treeFit, testData, type="class")
print(paste("CART Misclassification Rate:", sum(predictions!=testData[,1])/length(predictions)))
```

#### Problem 1C

```r
# MDA (1,1)
print(paste("MDA Misclassification Rate, Train (1,1): ", MDATrainer(train=trainData, test=trainData, subclasses=c(1,1))))
print(paste("MDA Misclassification Rate, Test (1,1): ", MDATrainer(train=trainData, test=testData, subclasses=c(1,1))))

# MDA (5,5)
""
print(paste("MDA Misclassification Rate, Train (5,5): ", MDATrainer(train=trainData, test=trainData, subclasses=c(5,5))))
print(paste("MDA Misclassification Rate, Test (5,5): ", MDATrainer(train=trainData, test=testData, subclasses=c(5,5))))

# MDA (10, 10)
""
print(paste("MDA Misclassification Rate, Train (10,10): ", MDATrainer(train=trainData, test=trainData, subclasses=c(10,10))))
print(paste("MDA Misclassification Rate, Test (10,10): ", MDATrainer(train=trainData, test=testData, subclasses=c(10,10))))
```

```r
# Visualizing trends with misclassification rate

maxSubclasses = 100

trainResults = list()
testResults = list()

for (i in 1:maxSubclasses) {
    trainResult = MDATrainer(train=trainData, test=trainData, subclasses=c(i,i))
    testResult = MDATrainer(train=trainData, test=testData, subclasses=c(i,i))
    
    trainResults = c(trainResults, trainResult)
    testResults = c(testResults, testResult)
}

x = 1:maxSubclasses
plot(x=x, y=trainResults)
plot(x=x, y=testResults)
```

#### Problem 1C

Based on our visualization, it appears that our model is obeying the bias-variance tradeoff. With smaller subclass sizes (and thus decreased complexity), we have greater bias but generally less variance. As we increase the model complexity, our training error drops dramatically but our test error increases due to overfitting.

```r
KNNTrainer = function(train, test, k) {
    predictions = knn(train=train[,-1], test=test[,-1], cl=train[,1], k=k)    
    return( sum(predictions!=test[,1])/length(predictions) )
}

# KNN (k=1)
print(paste("KNN with k=1, training set error misclassification:", 
            KNNTrainer(train=trainData, test=trainData, k=1)
           )
     )

print(paste("KNN with k=1, test set error misclassification:", 
            KNNTrainer(train=trainData, test=testData, k=1)
           )
     )

# KNN (k=5)
""
print(paste("KNN with k=5, training set error misclassification:", 
            KNNTrainer(train=trainData, test=trainData, k=5)
           )
     )

print(paste("KNN with k=5, test set error misclassification:", 
            KNNTrainer(train=trainData, test=testData, k=5)
           )
     )



# KNN (k=10)
""
print(paste("KNN with k=10, training set error misclassification:", 
            KNNTrainer(train=trainData, test=trainData, k=10)
           )
     )

print(paste("KNN with k=10, test set error misclassification:", 
            KNNTrainer(train=trainData, test=testData, k=10)
           )
     )
```

```r
# Visualizing trends with misclassification rate

maxK = 100

trainResults = list()
testResults = list()
for (i in 1:maxK) {
    trainResult = KNNTrainer(train=trainData, test=trainData, k=i)
    testResult = KNNTrainer(train=trainData, test=testData, k=i)
    
    trainResults = c(trainResults, trainResult)
    testResults = c(testResults, testResult)
}

plot(x=1:maxK, y=trainResults)
plot(x=1:maxK, y=testResults)
```

#### Problem 1D

So one very nonintuitive thing with KNN is that model complexity *increases* with smaller k's. That's part of the reason we see a training error of 0 with k=1 -- we evaluate all points of K and ask what the classification is of its single nearest neighbor. Since the nearest neighbor is always the point we're evaluating, the classes will always be accurate to the training set. Interestingly enough, kNN with large K's increases error rate on the training set but there's something very unusual going on with error rates on the test set. For some reason, we end up with just a few categories or unique error rates -- likely because models with the same error rate classify in much the same way despite differing k values and get the same few points wrong. I would've expected an increasing error rate with increasing K on the training set but since that doesn't happen, I'd need to do some more evaluation of why our error rate on testResults is acting the way it is.
